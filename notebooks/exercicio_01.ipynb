```python
import os
import sys
import glob
import cv2
import copy
import matplotlib.pyplot as plt
```

```python
PATH_IMGS = os.path.join('..', 'imagens')
imgs_list = glob.glob(PATH_IMGS+os.path.sep+'pais*')
```

```python
def plotar_canais(img, size=(20,15)):
    
    fig = plt.figure(figsize=size)
    labels = ['original', 'cinza', 'vermelho', 'verde', 'azul']
    
    img = cv2.imread(img)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    imgs_processadas = []
    imgs_processadas.append(img)
    imgs_processadas.append(img_gray)
    
    img_red = copy.copy(img)
    img_red[:,:,0] = 0
    img_red[:,:,1] = 0
    
    img_green = copy.copy(img)
    img_green[:,:,0] = 0
    img_green[:,:,2] = 0
    
    img_blue = copy.copy(img)
    img_blue[:,:,1] = 0
    img_blue[:,:,2] = 0
    
    imgs_processadas.extend([img_red, img_green, img_blue])

    for j in range(0, len(imgs_processadas)):
        plt.subplot(1, 5, j+1)
        plt.axis('off')
        plt.title(labels[j])
        img = cv2.cvtColor(imgs_processadas[j], cv2.COLOR_BGR2RGB)
        plt.imshow(img)
    plt.show()
```

```python
for img in imgs_list: 
    plotar_canais(img)
```
