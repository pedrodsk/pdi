```python
import os
import sys
import glob
import cv2
import copy
import matplotlib.pyplot as plt
```

```python
PATH_IMGS = os.path.join('..', 'imagens')
imgs_list = glob.glob(PATH_IMGS+os.path.sep+'pais*')
```

```python
def plotar_canais(img, size=(20,15)):
    
    fig = plt.figure(figsize=size)
    labels = ['original', 'sem canal azul', 'sem canal verde', 'sem canal vermelho']
    img = cv2.imread(img)
    
    imgs_processadas = []
    imgs_processadas.append(img)
    
    for i in range(0,3):
        img_no_channel = copy.copy(img)
        img_no_channel[:,:,i] = 0
        imgs_processadas.append(img_no_channel)

    for j in range(0, len(imgs_processadas)):
        plt.subplot(1, 4, j+1)
        plt.title(labels[j])
        img = cv2.cvtColor(imgs_processadas[j], cv2.COLOR_BGR2RGB)
        plt.imshow(img)
    plt.show()
```

```python
for img in imgs_list: 
    plotar_canais(img)
```
